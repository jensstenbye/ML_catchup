```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
```

```python
# Reshaping cells.
x = np.array([1, 2, 3, 4, 5, 6])
print('x before reshaping:\n', x)
x_reshape = x.reshape((2, 3)) # Reshape is a method
print('x after reshape: \n', x_reshape)
```

```python
# The reshaped array still points to the same array in memory, and thus changing an element
# also changes the corresponding element in x_reshaped
x[4] = 0
print('x after x[4] changed: \n', x)
print('x_reshape after x[4] changed: \n', x_reshape)
```

```python
print('x_reshape.shape returns shape of the array:\n', x_reshape.shape)
print('x_reshape.ndim returns #dimensions of the array:\n', x_reshape.ndim)
print('x_reshape.T returns transposed version of the array:\n', x_reshape.T)
```

## Stastical operations

```python
# Generate random normal distributed variables
x = np.random.normal(size=50) # default loc=50, scale=1
# Generate response variable by adding with points from another normal distribution
y = x + np.random.normal(loc=50, scale=1, size=50)

# Get the correlation matrix between the predictors and response (Highly correlated)
np.corrcoef(x, y)
```

```python
# set seed
rng = np.random.default_rng(1000)
```

```python
y = rng.standard_normal(10)
print('y:', y)
print('Var(y):', np.var(y)) #y.var()
print('Var(y) ddof1:', np.var(y, ddof=1))
print('Mean(y):', np.mean(y)) #y.mean()
print('std(y):', np.std(y)) #y.std()
```

```python
X = rng.standard_normal((10, 3))
print('X-Vertical mean:', X.mean(axis=0))
print('X-Horisontal mean:', X.mean(axis=1))
```

## Array manipulation

```python
print('linspace returns an sequence of value between a and b:\n', np.linspace(0, 10, 11))
print('arange returns a sequence spaced by a specific sted, default 1:\n', np.arange(0, 10, 2))
```

```python
A = np.array(np.arange(16)).reshape((4,4)) # Generate a 4x4 matrix of integers between 0 and 15
A
```

```python
print('A[1,2] fetches specific element at position:\n', A[1,2])
print('A[[1,2]] fetches specific rows:\n', A[[1,2]])
print('A[:,[1,2]] fetches specific columns:\n', A[:,[1,2]])
print('A[[0,2],[1,2]] fetches specific elements:\n', A[[0,2],[1,2]])
```

```python
# Generate submatrix made of specific rows and columns
# 1. Extract rows
# 2. Extract columns from extracted rows to get submatrix
A[[1,3]][:, [0,2]]
```

```python
# Generate submatrix using mesh object to index
idx = np.ix_([1,3],[0,2,3]) # Generates N array each with size 1 in all but the n-th dimension
print(idx)
print(A[idx])
```

```python
# Indexing using booleans
keep_rows = np.zeros(A.shape[0], dtype=bool)
print(keep_rows)
keep_rows[[1,3]]= True
print(keep_rows)
```

```python
print('Booleans rows for which their index is true:\n', A[keep_rows])
print('Integers returns the rows with index = integer:\n', A[np.array([0, 1, 0, 1])])
```

```python
# Use ix_() with booleans to index
keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True
idx_bool = np.ix_(keep_rows, keep_cols)
print('keep_rows:', keep_rows, '\nkeep_cols:', keep_cols)
print(idx_bool)
print(A[idx_bool])
```

## String formatting

```python
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5)) #Array size 127x5 standard normal distributed points
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape) #Array size 127x5 of da sampled from p1 with probability p2
A += M #Add array M to A
D = pd.DataFrame(A, columns=['food', 'bar', 'pickle', 'snack', 'popcorn']) #Load array A into DF with column names

D[:3]
```

```python
for col in D.columns:
    template = 'Column "{0}" has {1:.2%} missing values' #{1:.2%} insert variable 1 formated with 2 decimal digits
    print(template.format(col, np.isnan(D[col]).mean())) #% missing values in column
    #print(f'Column "{col}" has {np.isnan(D[col]).mean():.2%} missing values') #Can also be done with f-string
```
